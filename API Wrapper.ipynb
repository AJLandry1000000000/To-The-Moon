```python
# Full list of endpoints: https://www.alphavantage.co/documentation/
base_url = "https://www.alphavantage.co/query?"
api_key = '1PRBO66RYM7SV7B9'
```

```python
def _construct_url(*args, **kwargs): 
    param_string = ''
    for k in kwargs: 
        param_string += '&{}={}'.format(k, kwargs[k])
    return base_url + 'function={}'.format(args[0]) \
        + param_string + '&apikey={}'.format(api_key)
```

```python
def construct_ts_intraday(symbol, interval, **kwargs): 
    return _construct_url("TIME_SERIES_INTRADAY", symbol=symbol, interval=interval, **kwargs)

def construct_ts_daily_adjusted(symbol, **kwargs): 
    return _construct_url("TIME_SERIES_DAILY_ADJUSTED", symbol=symbol, **kwargs)
```

```python
sample_url = _construct_url("TIME_SERIES_DAILY_ADJUSTED", symbol="IBM", outputsize="full")

sample_url_2 = _construct_url("TIME_SERIES_INTRADAY", symbol="TSLA", interval="15min", outputsize="full")
```
